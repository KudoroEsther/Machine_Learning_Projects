```python
plt.figure(figsize=(15, 10))
for i, col in enumerate(['Va', 'Vb', 'Vc', 'Ia', 'Ib', 'Ic'], 1):
    plt.subplot(2, 3, i)
    sns.boxplot(x='Fault', y=col, data=df)
    plt.title(f'Distribution of {col} by Fault Class')
plt.tight_layout()
plt.savefig('Distribution.png', dpi=300, bbox_inches='tight')
plt.show()
```

```python
from sklearn.decomposition import PCA
from sklearn.manifold import TSNE

X = df[['Va', 'Vb', 'Vc', 'Ia', 'Ib', 'Ic']]
Y = df['Fault']

pca = PCA(n_components=2)
X_pca = pca.fit_transform(X)

plt.figure(figsize=(10, 7))
sns.scatterplot(x=X_pca[:, 0], y=X_pca[:, 1], hue=Y, palette='viridis', s=60)
plt.title('PCA: Feature Distribution by Fault Class')
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.legend(title='Fault')
plt.tight_layout()
plt.savefig('PCA.png', dpi=300, bbox_inches='tight')
plt.show()

tsne = TSNE(n_components=2, random_state=42, perplexity=30)
X_tsne = tsne.fit_transform(X)

plt.figure(figsize=(10, 7))
sns.scatterplot(x=X_tsne[:, 0], y=X_tsne[:, 1], hue=Y, palette='viridis', s=60)
plt.title('t-SNE: Feature Distribution by Fault Class')
plt.xlabel('t-SNE Component 1')
plt.ylabel('t-SNE Component 2')
plt.legend(title='Fault')
plt.tight_layout()
plt.savefig('t-sne.png', dpi=300, bbox_inches='tight')
plt.show()
```

```python
plt.figure(figsize=(15, 5))
phases = [('Va', 'Ia'), ('Vb', 'Ib'), ('Vc', 'Ic')]

for i, (v_col, i_col) in enumerate(phases, 1):
    plt.subplot(1, 3, i)
    sns.scatterplot(data=df, x=v_col, y=i_col, hue='Fault', palette='viridis', s=60)
    plt.title(f'{v_col} vs {i_col}')
    plt.grid(True, linestyle='--', alpha=0.7)
    
plt.tight_layout()
plt.savefig('vs.png', dpi=300, bbox_inches='tight')
plt.show()
```

```python
plt.figure(figsize=(15, 10))
features = ['Va', 'Vb', 'Vc', 'Ia', 'Ib', 'Ic']

for i, col in enumerate(features, 1):
    plt.subplot(2, 3, i)
    for fault in sorted(df['Fault'].unique()):
        subset = df[df['Fault'] == fault]
        sns.kdeplot(subset[col], label=f'Fault {fault}', fill=True, alpha=0.3)
    plt.title(f'Distribution of {col} by Fault Class')
    plt.grid(True, linestyle='--', alpha=0.7)
    plt.legend(title='Fault')

plt.tight_layout()
plt.savefig('dd.png', dpi=300, bbox_inches='tight')
plt.show()
```

```python
x=df.drop(['Fault'],axis=1)
y=df['Fault']
```

```python
x
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.30)
```

```python
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
x_train_scaled=scaler.fit_transform(x_train)
x_test_scaled=scaler.transform(x_test)
```

```python
from sklearn.metrics import classification_report,accuracy_score,confusion_matrix,ConfusionMatrixDisplay
from sklearn.ensemble import AdaBoostClassifier
model = AdaBoostClassifier()
model.fit(x_train_scaled,y_train)
y_pred=model.predict(x_test_scaled)
cm = confusion_matrix(y_test, y_pred,normalize='true')
print('\nClassification Report:\n',classification_report(y_test,y_pred))
print('\nAccuracy Score:\n',accuracy_score(y_test,y_pred))
disp = ConfusionMatrixDisplay(confusion_matrix=cm, 
                              display_labels=['No Fault (0)', 'Fault Type 1', 'Fault Type 2', 'Fault Type 3'])
disp.plot(cmap=plt.cm.Blues, values_format='.1%')
plt.title('Fault Classification Confusion Matrix (Normalized)', fontsize=16)
plt.xticks(rotation=25)
plt.show()
```

```python
from sklearn.svm import SVC
model=SVC()
model.fit(x_train_scaled,y_train)
y_pred=model.predict(x_test_scaled)
cm = confusion_matrix(y_test, y_pred,normalize='true')
print('\nClassification Report:\n',classification_report(y_test,y_pred))
print('\nAccuracy Score:\n',accuracy_score(y_test,y_pred))
disp = ConfusionMatrixDisplay(confusion_matrix=cm, 
                              display_labels=['No Fault (0)', 'Fault Type 1', 'Fault Type 2', 'Fault Type 3'])
disp.plot(cmap=plt.cm.Blues, values_format='.1%')
plt.title('Fault Classification Confusion Matrix (Normalized)', fontsize=16)
plt.xticks(rotation=25)
plt.show()
```

```python
from sklearn.tree import DecisionTreeClassifier
model = DecisionTreeClassifier()
model.fit(x_train_scaled,y_train)
y_pred=model.predict(x_test_scaled)
cm = confusion_matrix(y_test, y_pred,normalize='true')
print('\nClassification Report:\n',classification_report(y_test,y_pred))
print('\nAccuracy Score:\n',accuracy_score(y_test,y_pred))
disp = ConfusionMatrixDisplay(confusion_matrix=cm, 
                              display_labels=['No Fault (0)', 'Fault Type 1', 'Fault Type 2', 'Fault Type 3'])
disp.plot(cmap=plt.cm.Blues, values_format='.1%')
plt.title('Fault Classification Confusion Matrix (Normalized)', fontsize=16)
plt.xticks(rotation=25)
plt.show()
```
